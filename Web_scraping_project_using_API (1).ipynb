### Scraping multiple pages to csv file

```python
# for multiple pages
import requests

headers = {
    'authority': 'www.ebooks.com',
    'accept': '*/*',
    'accept-language': 'en-US,en;q=0.9',
    'content-type': 'application/json',
    'cookie': '.AspNetCore.Antiforgery.u9wWfcCnN54=CfDJ8BlKYisbCxpPrDydqdChgSpribbJllz-HIPQELYcG6dLiXVMaRl8s1ZVhLen922-3k1H6UskEv0Oterh5lBNpHf896vZoJh_tHsu156kUB04EEIGFyun0D-6Jok-iS0kyt-o8kLQuOIf9JjI2ljjCgg; __cflb=02DiuEnpWqjzEfhLpxXo57HnJ9k24HkZrzUUDAcTRo5rk; _ga=GA1.1.420928295.1699251755; _ga_FQMQQ21J7D=GS1.1.1699292218.2.1.1699293983.15.0.0; _uetsid=e00ed1007c6c11ee92fe67c932c4bb07; _uetvid=e00f02007c6c11eea59883d809a8039d',
    'referer': 'https://www.ebooks.com/en-in/subjects/fantasy/?pageNumber=3',
    'sec-ch-ua': '^\\^Chromium^\\^;v=^\\^118^\\^, ^\\^Google',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '^\\^Windows^\\^',
    'sec-fetch-dest': 'empty',
    'sec-fetch-mode': 'cors',
    'sec-fetch-site': 'same-origin',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36',
}

title = []
subtitle = []
author = []
publisher = []
publication_yr = []
price = []


for i in range(1,11):

  params = (
      ('pageNumber', str(i)),
      ('CountryCode', 'IN'),
      ('subjectId', '329'),
  )

  # creating response
  response = requests.get('https://www.ebooks.com/api/search/subject/', headers=headers, params=params)

  # creating json object for response
  res_json = response.json()

  # extracting the books key from created json object
  res_books = res_json['books']

  for result in res_books:
    title.append(result['title'])
    subtitle.append(result['subtitle'])
    author.append(result['authors'][0]['name'])
    publisher.append(result['publisher'])
    publication_yr.append(result['publication_year'])
    price.append(result['price'])

```

```python
book_df = pd.DataFrame({'title':title, 'subtitle':subtitle, 'author':author, 'publisher':publisher, 'publication_yr':publication_yr, 'price':price})
```
